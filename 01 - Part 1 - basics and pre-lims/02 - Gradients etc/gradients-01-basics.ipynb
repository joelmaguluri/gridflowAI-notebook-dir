```python
import numpy as np
from IPython.display import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Get 100 evenly spaced numbers between -1 and 3 ( chosen to ensure steep curve)
x = np.linspace(-2, 2, 2000)
```

```python
# say the given function is 
f = x**3 - x 
```

```python
# Plot the curve
plt.figure(figsize=(12, 6))

plt.grid()

#plt.axis([-15, 7, -5, 15])

plt.axhline(y=0, color='r')
plt.axvline(x=0, color='c')

plt.plot(x, f)
plt.show()
```

```python
# Define the quadratic function
def quadratic_function(x):
    return x**2 - 3*x + 2

# Define the derivative of the quadratic function
def derivative_function(x):
    return 2*x - 3
```

```python
# Generate x values for plotting
x = np.linspace(-2, 4, 100)

# Compute y values for the quadratic function
y = quadratic_function(x)
```

```python
# Compute tangent lines
tangent_lines_x = [0.5, 1.5, 2.5]
tangent_lines_y = [quadratic_function(x_val) + derivative_function(x_val) * (x - x_val) for x_val in tangent_lines_x]
```

```python
# Plot the quadratic function
plt.plot(x, y, label='Quadratic Function')

# Plot tangent lines
for i in range(len(tangent_lines_x)):
    plt.plot(x, tangent_lines_y[i], label=f'Tangent at x={tangent_lines_x[i]}')

# Set plot title and labels
plt.title('Quadratic Function and Tangent Lines')
plt.xlabel('x')
plt.ylabel('y')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

our function started out with positive slope (somewhere along the negative x axis). 

This tells us that it must have grown for some time. (Notice how the tangent lines are pointing towards increasing y values when their slope is positive. That's what we mean when we say that the function must have been growing). 

Eventually, it flattened out, and precisely at the point where the slope of its tangent was zero, the function had a little "high". (We will call that place a local maximum.) 

After that, our function had a negative slope. This tells us it was decreasing. After some time, the slope flattened out to zero and the function had a local minimum. 

After that it just grew and grew. 

We can summarize our findings as follows:

    A positive derivative means that the function is increasing
    A negative derivative means that the function is decreasing
    A zero derivative means that the function has some special behaviour at the given point. It may have a local maximum, a local minimum, (or in some cases, as we will see later, a "turning" point)

As a last remark we should remember that the derivative of a function is, itself, a function since it varies from point to point.

#### how do we compute gradients

**Example 1**

```python
import sympy as sp

# Define the variables and the function
x, y = sp.symbols('x y')
f = x**2 + 2*y**3

# Compute the gradient
grad = [sp.diff(f, var) for var in (x, y)]

# Print the gradient
print("Gradient:", grad)
```

In this example, we define two variables `x` and `y` using `sympy.symbols()`. Then, we define a function f that represents a scalar-valued function of x and y.

Next, we use a list comprehension to compute the `partial derivatives` of f with respect to each variable. The `sympy.diff()` function is used to compute the derivatives.

Finally, we print the computed gradient, which is a list of the partial derivatives [2*x, 6*y**2].

**Example 2**

```python
# Compute the derivative of a simple polynomial function
x = sp.Symbol('x')

f = x**3 + 2*x**2 - 5*x + 1

f_prime = sp.diff(f, x)

print("Derivative of f(x):", f_prime)
```

**Example 3**

```python
# Compute the derivative of a trigonometric function
x = sp.Symbol('x')
g = sp.sin(x) + sp.cos(x)
g_prime = sp.diff(g, x)
print("Derivative of g(x):", g_prime)
```

**Example 4**

```python
# Compute the partial derivatives of a multivariable function
x, y = sp.symbols('x y')
h = x**2 * sp.sin(y) + y**2 * sp.cos(x)
h_x = sp.diff(h, x)
h_y = sp.diff(h, y)
print("Partial derivative of h(x, y) with respect to x:", h_x)
print("Partial derivative of h(x, y) with respect to y:", h_y)
```
